## <font color=#33A5FF>****2. Funciones básicas de lectura de ficheros****<tfont>

**index_col** = Cambiar el índice

```python
# Opción index_col
df = pd.read_excel('titanic.xlsx', engine='openpyxl', index_col=5)
# Seleccionamos el indice por su posicion, empezando por el 0 desde la izda.
df
```

**nrows** = utilizar las n primeras filas

```python
# Opción nrows
df = pd.read_excel('titanic.xlsx', engine='openpyxl', nrows=5)
df
```

**usecols**: Utilizar columnas específicas

```python
# Opción usecols
# Le pasamos una lista con los nombres de las coumnas que queremos incluir en el df
df = pd.read_excel('titanic.xlsx', engine='openpyxl', usecols=['Nombre', 'Sexo', 'Edad'])
df.head(10)
```

**skiprows**: saltarse filas de abajo hacia arriba

```python
# Opción skiprows
# Se le pasa un entero. Salta las filas que le pases. OJO, la fila 1 son los encabezados.
df = pd.read_excel('titanic.xlsx', engine='openpyxl', skiprows=3)
df 
```

**sheet_name**: Seleccionar las pestañas del libro excel

```python
# Opción sheet_name
# Los files xlsx pueden contener varias hojas excell. Para acceder a ellas se usa esta funcion
# Le puedes pasar un numero de indice o el nombre de la hoja
df = pd.read_excel('titanic.xlsx', engine='openpyxl', sheet_name=1) # sheet_name='Pasajeros_2'
df
```

**header**: Se utiliza cuando las columnas no tienen encabezado

```python
# Opción header
df = pd.read_excel('titanic.xlsx', engine='openpyxl') # ,header = None, lista = ['nombre', 'sexo']
# Sirve para pasarle una lista con los nombres que quieres para los encabezados, como complemento de skiprows que nos
# devuelve el df sin encabezado
```
